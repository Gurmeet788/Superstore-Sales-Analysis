```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/superstore-dataset-final/Sample - Superstore.csv', encoding='ISO-8859-1')
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['unit price'] = df['Sales']/df['Quantity']
```

```python
product_by = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False)
```

```python
product_by
```

```python
sales_region = df.groupby('Region')['Sales'].sum().sort_values(ascending=True)
```

```python
sales_region
```

```python
top_5 = product_by.head(5)
```

```python
top_5
```

```python
df['Monthes'] = df['Order Date'].dt.to_period('M')
```

```python
Months_sales = df.groupby('Monthes')['Sales'].sum()
```

```python
Months_sales
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6)) 
top_5.plot(kind='bar', color='skyblue')
plt.title("Top 5 Product")
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45,ha='right') 
plt.tight_layout() 
```


**create a figuare with size 10 by 6plot function is direclty connected with matplotlib in backed ha mean horizonatal plt.tight_layout() overlapping saver**


```python
plt.figure(figsize = (10,6))
sales_region.plot(kind = 'pie',autopct = '%1.1f%%' ,startangle = 180)
plt.title("sales by region")
plt.tight_layout()
```

```python
plt.figure(figsize=(10,10))
Months_sales.plot(kind = 'line', marker ='p',linestyle = '--',color = 'green')
plt.title('Monthly sale')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid(True)
plt.tight_layout()
```

```python
import numpy as np
```

```python
Months_sales_ma = Months_sales.rolling(window = 3).mean()
```

```python
plt.figure(figsize=(10,6))
Months_sales.plot(label='Monthly Sales', marker='o')
Months_sales_ma.plot(label = '3 mothes Avg', linestyle = '--', marker = '^')
plt.title('Moving Avrage')
plt.xlabel('Mothes')
plt.ylabel('sales')
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.xticks(rotation = 45, ha = 'right')
```

```python
monthly_sales_product = df.groupby(['Monthes','Product Name'])['Sales'].sum()
```

```python
best_selling_product = monthly_sales_product.groupby('Monthes').idxmax() #indmax return max value indx
```

** **#indmax return max indx with value if we work as value it will work as value and if we pass in loc function it will work as indx**

```python
best_product = monthly_sales_product.loc[best_selling_product]
```

**loc function take paremter indx and best_selling_product return max value indx mean where max value present from monthly_sales_product**

```python
best_product
```

```python
best_selling_product.sort_index(ascending = True)
```

**indmax as value**

```python
plt.figure(figsize=(10,6))
best_product.plot(kind = 'line',marker = 'o', linestyle = '--',color = 'green')
plt.title('Best monthly_sales_product')
plt.xticks(rotation = 45, ha = 'right')
plt.grid(True)
plt.xlabel('Monthes'+'Best Product Name')
plt.ylabel('Sales')
```
